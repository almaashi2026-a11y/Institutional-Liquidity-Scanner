```python

"""
PRE-PUMP SCANNER v2 - رصد التجميع المؤسسي قبل الضخ (احترافي)
"""

!pip install yfinance --upgrade --quiet

import yfinance as yf
import pandas as pd
import numpy as np
import requests
from concurrent.futures import ThreadPoolExecutor, as_completed
from datetime import datetime
import time
import warnings
warnings.filterwarnings("ignore")
pd.set_option("display.max_rows", 100)
pd.set_option("display.width", 160)
pd.set_option("display.colheader_justify", "center")

PRICE_MIN = 0.20
PRICE_MAX = 10.00

MIN_AVG_VOLUME = 50_000
RVOL_MIN = 1.5
RVOL_MAX_FOR_QUIET = 4.0
MAX_DAILY_CHANGE_FOR_QUIET = 6.0

MIN_ACCUMULATION_FOR_READY = 55
MIN_EXPLOSION_PROB_FOR_READY = 45

MAX_WORKERS = 20
BATCH_SIZE = 50

TELEGRAM_BOT_TOKEN = "8524780143:AAHwM-h4Pi7taJNVQZvLbma-5Id9BaJ5X6Y"
TELEGRAM_CHAT_ID = "7990990947"


def get_nasdaq_symbols():
    url = "https://www.nasdaqtrader.com/dynamic/SymDir/nasdaqlisted.txt"
    try:
        df = pd.read_csv(url, sep="|")
        df = df[df["Test Issue"] == "N"]
        symbols = df["Symbol"].dropna().tolist()
        symbols = [s for s in symbols if isinstance(s, str) and s.isalpha()]
        print(f"✅ تم جلب {len(symbols)} رمز من NASDAQ")
        return symbols
    except Exception as e:
        print(f"⚠️ فشل جلب القائمة الكاملة: {e}")
        return []


def send_telegram_alert(message):
    if TELEGRAM_BOT_TOKEN == "ضع_التوكن_هنا":
        return False
    try:
        url = f"https://api.telegram.org/bot{TELEGRAM_BOT_TOKEN}/sendMessage"
        response = requests.post(url, data={
            "chat_id": TELEGRAM_CHAT_ID,
            "text": message,
            "parse_mode": "HTML"
        }, timeout=5)
        return response.status_code == 200
    except Exception:
        return False


def test_telegram_connection():
    """يرسل رسالة تجريبية للتأكد إن البوت شغال ومتصل صحيح"""
    print("📡 جاري اختبار الاتصال بالبوت...")
    test_message = (
        f"✅ <b>البوت شغال!</b>\n\n"
        f"السكانر متصل بنجاح بتيليجرام.\n"
        f"⏰ {datetime.now().strftime('%Y-%m-%d %H:%M:%S')}"
    )
    success = send_telegram_alert(test_message)
    if success:
        print("✅ تم إرسال رسالة الاختبار بنجاح - تحقق من تيليجرام")
    else:
        print("❌ فشل الاتصال بالبوت - تحقق من التوكن أو Chat ID")
    return success


def calculate_cvd_approx(df):
    if len(df) < 2:
        return pd.Series([0] * len(df))
    money_flow_multiplier = ((df["Close"] - df["Low"]) - (df["High"] - df["Close"])) / \
                            (df["High"] - df["Low"]).replace(0, np.nan)
    money_flow_multiplier = money_flow_multiplier.fillna(0)
    money_flow_volume = money_flow_multiplier * df["Volume"]
    return money_flow_volume.cumsum()


def calculate_atr(df, period=14):
    high_low = df["High"] - df["Low"]
    high_close = (df["High"] - df["Close"].shift()).abs()
    low_close = (df["Low"] - df["Close"].shift()).abs()
    tr = pd.concat([high_low, high_close, low_close], axis=1).max(axis=1)
    return tr.rolling(window=period).mean()


def calculate_vwap(df):
    typical_price = (df["High"] + df["Low"] + df["Close"]) / 3
    return (typical_price * df["Volume"]).cumsum() / df["Volume"].cumsum()


def calculate_bb_width(df, period=20, std=2):
    sma = df["Close"].rolling(window=period).mean()
    rolling_std = df["Close"].rolling(window=period).std()
    upper = sma + (rolling_std * std)
    lower = sma - (rolling_std * std)
    return (upper - lower) / sma


def analyze_symbol(symbol):
    try:
        ticker = yf.Ticker(symbol)

        df_daily = ticker.history(period="1mo", interval="1d")
        if df_daily.empty or len(df_daily) < 15:
            return None

        last_price = df_daily["Close"].iloc[-1]
        if not (PRICE_MIN <= last_price <= PRICE_MAX):
            return None

        avg_volume_20d = df_daily["Volume"].iloc[-20:].mean()
        if avg_volume_20d < MIN_AVG_VOLUME:
            return None

        df_intraday = ticker.history(period="5d", interval="15m")
        if df_intraday.empty or len(df_intraday) < 10:
            return None

        today_volume = df_daily["Volume"].iloc[-1]
        rvol = today_volume / avg_volume_20d if avg_volume_20d > 0 else 0

        open_price = df_daily["Open"].iloc[-1]
        daily_change_pct = ((last_price - open_price) / open_price) * 100 if open_price > 0 else 0

        atr = calculate_atr(df_daily)
        atr_now = atr.iloc[-1]
        atr_avg = atr.iloc[-10:].mean()
        compression_ratio = atr_now / atr_avg if atr_avg > 0 else 1

        cvd = calculate_cvd_approx(df_intraday)
        cvd_trend = 0
        cvd_strength = 0
        if len(cvd) >= 10:
            recent_cvd = cvd.iloc[-10:]
            cvd_slope = np.polyfit(range(len(recent_cvd)), recent_cvd.values, 1)[0]
            cvd_trend = cvd_slope
            cvd_range = recent_cvd.max() - recent_cvd.min()
            cvd_strength = abs(cvd_slope) / (cvd_range / 10) if cvd_range > 0 else 0

        vwap = calculate_vwap(df_intraday)
        vwap_now = vwap.iloc[-1]
        price_vs_vwap_pct = ((last_price - vwap_now) / vwap_now) * 100 if vwap_now > 0 else 0

        bb_width = calculate_bb_width(df_daily)
        bb_width_now = bb_width.iloc[-1]
        bb_width_avg = bb_width.iloc[-20:].mean()
        bb_compression_ratio = bb_width_now / bb_width_avg if bb_width_avg > 0 else 1

        high_10d = df_daily["High"].iloc[-10:].max()
        distance_from_high_pct = ((high_10d - last_price) / high_10d) * 100 if high_10d > 0 else 0

        recent_closes = df_intraday["Close"].iloc[-5:]
        price_momentum = np.polyfit(range(len(recent_closes)), recent_closes.values, 1)[0] if len(recent_closes) >= 3 else 0
        price_momentum_pct = (price_momentum / last_price) * 100 if last_price > 0 else 0

        acc_score = 0

        if rvol >= RVOL_MIN and abs(daily_change_pct) <= MAX_DAILY_CHANGE_FOR_QUIET:
            acc_score += min(30, (min(rvol, RVOL_MAX_FOR_QUIET) / RVOL_MAX_FOR_QUIET) * 30)

        if cvd_trend > 0:
            cvd_pts = min(35, cvd_strength * 35) if abs(daily_change_pct) <= MAX_DAILY_CHANGE_FOR_QUIET else min(20, cvd_strength * 20)
            acc_score += cvd_pts

        if 0 <= price_vs_vwap_pct <= 3:
            acc_score += 20
        elif price_vs_vwap_pct > 3:
            acc_score += 10

        if bb_compression_ratio < 0.85:
            acc_score += 15
        elif bb_compression_ratio < 1.0:
            acc_score += 7

        acc_score = round(min(acc_score, 100), 1)

        exp_score = 0

        if compression_ratio < 1.0:
            exp_score += min(35, (1 - compression_ratio) * 70)

        if distance_from_high_pct <= 15:
            exp_score += (15 - min(distance_from_high_pct, 15)) / 15 * 30
            if last_price >= high_10d:
                exp_score += 30

        if price_momentum_pct > 0:
            exp_score += min(20, price_momentum_pct * 10)

        if rvol >= 2.0:
            exp_score += 15

        exp_score = round(min(exp_score, 100), 1)

        if acc_score >= 75 and exp_score >= 75:
            entry_quality = "🟢 دخول مثالي"
        elif acc_score >= MIN_ACCUMULATION_FOR_READY and exp_score >= MIN_EXPLOSION_PROB_FOR_READY:
            entry_quality = "🟢 جاهز للدخول"
        elif acc_score >= MIN_ACCUMULATION_FOR_READY and exp_score < MIN_EXPLOSION_PROB_FOR_READY:
            entry_quality = "🟡 تجميع قوي - بدري"
        elif acc_score < MIN_ACCUMULATION_FOR_READY and exp_score >= MIN_EXPLOSION_PROB_FOR_READY:
            entry_quality = "🟠 زخم بلا تجميع - حذر"
        else:
            entry_quality = "⚪ ضعيف"

        is_ready = (acc_score >= MIN_ACCUMULATION_FOR_READY) and (exp_score >= MIN_EXPLOSION_PROB_FOR_READY)

        return {
            "الرمز": symbol,
            "السعر": round(last_price, 3),
            "قوة_التجميع_%": acc_score,
            "احتمالية_الانفجار_%": exp_score,
            "تقييم_الدخول": entry_quality,
            "RVOL": round(rvol, 2),
            "التغير_%": round(daily_change_pct, 2),
            "بعد_عن_القمة_%": round(distance_from_high_pct, 2),
            "ضغط_ATR": round(compression_ratio, 2),
            "is_ready": is_ready,
        }

    except Exception:
        return None


def run_scan(symbols, batch_size=BATCH_SIZE, max_workers=MAX_WORKERS):
    results = []
    total = len(symbols)

    for i in range(0, total, batch_size):
        batch = symbols[i:i + batch_size]
        with ThreadPoolExecutor(max_workers=max_workers) as executor:
            futures = {executor.submit(analyze_symbol, sym): sym for sym in batch}
            for future in as_completed(futures):
                result = future.result()
                if result is not None:
                    results.append(result)

        done = min(i + batch_size, total)
        print(f"\r🔍 تم فحص {done}/{total} رمز...", end="", flush=True)
        time.sleep(0.01)

    print()
    return results


def display_ready_table(results):
    ready = [r for r in results if r["is_ready"]]

    print("\n" + "=" * 90)
    print(f"📊 فحص Pre-Pump - {datetime.now().strftime('%Y-%m-%d %H:%M:%S')}")
    print(f"إجمالي الأسهم المفحوصة المطابقة للشروط الأساسية: {len(results)}")
    print("=" * 90)

    if not ready:
        print("⚠️ لا توجد أسهم جاهزة للدخول الفوري في هذا المسح")
        print("=" * 90)
        return []

    df = pd.DataFrame(ready)
    df = df.drop(columns=["is_ready"])
    df = df.sort_values(by=["قوة_التجميع_%", "احتمالية_الانفجار_%"])
    df = df.reset_index(drop=True)
    df.index = df.index + 1

    print(f"\n🔥 الأسهم الجاهزة للدخول الفوري: {len(ready)}\n")
    print(df.to_string())
    print("\n" + "=" * 90)

    return ready


def send_alerts(ready_results):
    for r in ready_results:
        message = (
            f"🔥 <b>Pre-Pump - جاهز للدخول</b>\n\n"
            f"📌 الرمز: <b>{r['الرمز']}</b>\n"
            f"💰 السعر: ${r['السعر']}\n"
            f"📊 قوة التجميع: {r['قوة_التجميع_%']}%\n"
            f"💥 احتمالية الانفجار: {r['احتمالية_الانفجار_%']}%\n"
            f"⭐ التقييم: {r['تقييم_الدخول']}\n"
            f"📈 RVOL: {r['RVOL']}x\n"
            f"📉 التغير اليومي: {r['التغير_%']}%\n"
            f"🎯 بعد عن القمة: {r['بعد_عن_القمة_%']}%\n\n"
            f"⏰ {datetime.now().strftime('%H:%M:%S')}"
        )
        send_telegram_alert(message)


def main(continuous=False, interval_minutes=15):
    symbols = get_nasdaq_symbols()
    if not symbols:
        print("❌ لم يتم جلب أي رموز، تأكد من الاتصال بالإنترنت")
        return

    if not continuous:
        results = run_scan(symbols)
        ready = display_ready_table(results)
        if ready:
            send_alerts(ready)
    else:
        print(f"🔄 وضع التشغيل المستمر - فحص كل {interval_minutes} دقيقة")
        while True:
            try:
                results = run_scan(symbols)
                ready = display_ready_table(results)
                if ready:
                    send_alerts(ready)
                print(f"\n⏳ الانتظار {interval_minutes} دقيقة للفحص التالي...\n")
                time.sleep(interval_minutes * 60)
            except KeyboardInterrupt:
                print("\n🛑 تم إيقاف السكانر")
                break
            except Exception as e:
                print(f"⚠️ خطأ: {e} - إعادة المحاولة بعد دقيقة")
                time.sleep(60)


main(continuous=False)

# main(continuous=True, interval_minutes=15)
```
